# Perceptron

Perceptronというのは`複数の信号`を入力として受け取り、`ひとつの信号`を出力することです。 以下の図で、x1, x2は入力信号`input` yは出力信号、w1, w2は`Weight`を表示します。

スクリーンショット 2020-11-29 3.18.15.png

## weights and bias

- weights (w1, w2)は入力信号への重要度をコントロールするパラメータ
- biasは発火のしやすさ


### Implement a Perceptron.

- AND
- NAND
- OR
- XOR

```
import numpy as np
```

### AND

```
def AND(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([0.5, 0.5])
  bias = - 0.7

  tmp = np.sum(weights * input) + bias

  if tmp <= 0:
    return 0
  else:
    return 1
```

```
print("input data 0, 0 : {}".format(AND(0, 0)))
print("input data 0, 1 : {}".format(AND(0, 1)))
print("input data 1, 0 : {}".format(AND(1, 0)))
print("input data 1, 1 : {}".format(AND(1, 1)))
```

### NAND

```
def NAND(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([-0.5, -0.5])
  bias = 0.7

  tmp = np.sum(weights * input) + bias

  if tmp <= 0 :
    return 0
  else:
    return 1
```

```
print("input data 0, 0 : {}".format(NAND(0, 0)))
print("input data 0, 1 : {}".format(NAND(0, 1)))
print("input data 1, 0 : {}".format(NAND(1, 0)))
print("input data 1, 1 : {}".format(NAND(1, 1)))
```

### OR


```
def OR(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([0.5, 0.5])
  bias = -0.2

  tmp = np.sum(weights * input) + bias

  if tmp <= 0:
    return 0
  else:
    return 1
```

```
print("input data 0, 0 : {}".format(OR(0, 0)))
print("input data 0, 1 : {}".format(OR(0, 1)))
print("input data 1, 0 : {}".format(OR(1, 0)))
print("input data 1, 1 : {}".format(OR(1, 1)))
```

## Perceptronの限界

### Exclusive

Exclusiveは（排他的論理和）x1, x2どちらの値が`1`の場合のみ結果を`1`として出力する

スクリーンショット 2020-11-29 11.12.16.png

### XOR

Exclusiveを表現するにはAND、NAND、ORを利用することで可能になる

```
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```
print("input data 0, 0 : {}".format(XOR(0, 0)))
print("input data 0, 1 : {}".format(XOR(0, 1)))
print("input data 1, 0 : {}".format(XOR(1, 0)))
print("input data 1, 1 : {}".format(XOR(1, 1)))
```
